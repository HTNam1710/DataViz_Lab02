```python
# merge các bảng
data = pd.concat([AGR, BID, CTG, TCB, VCB]).reset_index(drop=True)
data
```

```python
# kiểm tra kiểu dữ liệu
data.dtypes
```

```python
# chuyển đổi kiểu dữ liệu
data['Date'] = pd.to_datetime(data['Date'])
for col in ['Price', 'Open', 'High', 'Low']:
    data[col] = data[col].str.replace(',', '').astype(float)

def volume_to_float(x):
    if 'K' in x:
        return float(x.replace('K', '')) * 1e3
    elif 'M' in x:
        return float(x.replace('M', '')) * 1e6
    return float(x)

data['Vol.'] = data['Vol.'].apply(volume_to_float)

data['Change %'] = data['Change %'].str.rstrip('%').astype(float)

data.dtypes
```

```python
data.set_index('Date', inplace=True)
data
```

## Ý nghĩa các cột

- *Date:* Ngày giao dịch diễn ra.

- *Price:* Giá đóng cửa của cổ phiếu vào ngày giao dịch.
  
- *Open:* Giá mở cửa của cổ phiếu vào ngày giao dịch.
  
- *High:* Giá cao nhất của cổ phiếu trong ngày giao dịch.
  
- *Low:* Giá thấp nhất của cổ phiếu trong ngày giao dịch.
  
- *Vol. (Volume):* Khối lượng giao dịch của cổ phiếu trong ngày.
  
- *Change %:* Phần trăm thay đổi của giá cổ phiếu so với ngày giao dịch trước đó.
  
- *XSTC:* Mã cổ phiếu của công ty, với các giá trị có thể là 'AGR', 'BID', 'CTG', 'TCB', 'VCB'.

```python
# Do thời gian chưa sắp xếp nên ta sắp xếp lại
def sort_by_stock_and_date(df):
    """
    Sắp xếp DataFrame theo mã cổ phiếu ('XSTC') và sau đó theo ngày ('Date').
    
    Args:
    - df: DataFrame chứa dữ liệu cần sắp xếp.

    Returns:
    - DataFrame đã sắp xếp.
    """
    # Sắp xếp DataFrame theo cột 'XSTC' và 'Date'
    sorted_df = df.sort_values(by=['XSTC', 'Date'], ascending=[True, True])
    return sorted_df

data = sort_by_stock_and_date(data)
data
```
